```python
import bs4
import requests
import re

count = 0
html_page = requests.get("http://www.mospi.gov.in")
soup = bs4.BeautifulSoup(html_page.text,'lxml')
links = set()
files = set()
links_copy = set()

for link in soup.findAll('a', attrs={'href': re.compile("^http://www.mospi.gov.in")}):
    if (re.findall('.PDF$',link['href'])) or (re.findall('.pdf$',link['href'])):
        files.add(link.get('href'))
    else:
        links.add(link.get('href'))
        
for link in soup.findAll('a', attrs={'href': re.compile("^/+")}):
    if (re.findall('.PDF$',link['href'])) or (re.findall('.pdf$',link['href'])):
        files.add("http://www.mospi.gov.in"+link.get('href'))
    else:
        links.add("http://www.mospi.gov.in"+link.get('href'))

for x in links:
    print(x)
```
